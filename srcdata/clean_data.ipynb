```python
import pandas as pd

pd.options.display.max_columns = None
```

#### Read Data

```python
input_data = pd.read_csv("SumitroDatta/PlayerPerGame.csv")
input_data = input_data.sort_values('player_id')

input_data.head()
```

#### Clean Data

```python
renames = {'seas_id' : 'SeasonID', 
           'season' : 'Season', 
           'player_id' : 'PlayerID', 
           'player' : 'Name', 
           'birth_year' : 'BirthYear', 
           'pos' : 'Position', 
           'age' : 'Age', 
           'experience' : 'YearsExperience', 
           'lg' : 'League', 
           'tm' : 'Team', 
           'g' : 'Games', 
           'gs' : 'GamesStarted', 
           'mp_per_game' : 'MinutesPlayedPerGame', 
           'fg_per_game' : 'FieldGoalsPerGame', 
           'fga_per_game' : 'FieldGoalAttemptsPerGame', 
           'fg_percent' : 'FieldGoalPercent', 
           'x3p_per_game' : 'x3PointsPerGame', 
           'x3pa_per_game' : 'x3PointAttemptsPerGame', 
           'x3p_percent' : 'x3PointPercent', 
           'x2p_per_game' : 'x2PointsPerGame', 
           'x2pa_per_game' : 'x2PointAttemptsPerGame', 
           'x2p_percent' : 'x2PointPercent', 
           'e_fg_percent' : 'EffectiveFieldGoalPercent', 
           'ft_per_game' : 'FreeThrowsPerGame', 
           'fta_per_game' : 'FreeThrowAttemptsPerGame', 
           'ft_percent' : 'FreeThrowPercent', 
           'orb_per_game' : 'OffensiveReboundsPerGame', 
           'drb_per_game' : 'DefensiveReboundsPerGame', 
           'trb_per_game' : 'ReboundsPerGame', 
           'ast_per_game' : 'AssistsPerGame', 
           'stl_per_game' : 'StealsPerGame', 
           'blk_per_game' : 'BlocksPerGame', 
           'tov_per_game' : 'TakeoversPerGame', 
           'pf_per_game' : 'PersonalFoulsPerGame', 
           'pts_per_game' : 'PointsPerGame'
        }

rdata = input_data.rename(columns=renames)
#rdata['BirthYear'] = pd.to_numeric(rdata['BirthYear'], downcast='integer', errors='coerce')
#rdata['Age'] = pd.to_numeric(rdata['Age'], downcast='integer', errors='coerce')
rdata = rdata.astype({'BirthYear' : 'Int64', 'Age' : 'Int64'}, errors='ignore')
rdata.head()
```

```python
player_info = rdata[['PlayerID', 'Name', 'BirthYear']]
display(player_info[:5])
```

```python
column_info = pd.DataFrame({
    'ColumnNames' : list(rdata.columns),
    'DataType' : list(rdata.dtypes)
})
column_info.transpose().head()
```

```python
season_stats = rdata[['Season', 'PlayerID', 'Age', 'YearsExperience', 'League', 'Team', 'Position', 'Games', 'PointsPerGame', 'AssistsPerGame', 'BlocksPerGame', 'ReboundsPerGame', 'EffectiveFieldGoalPercent']]
display(season_stats[:5])
```

```python
player_info.to_csv('ReadyData/PlayerInfo.csv', index=False)
```

```python
season_stats.to_csv('ReadyData/PlayerSeasonStats.csv', index=False)
```
